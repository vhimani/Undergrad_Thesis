```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('graphs.csv')
df.head()
df['year'] = df['year'].astype(str)
```

```python
df_all = df[['year','work paper all', 'leisure paper all', 'home paper all', 'childcare all']]
df_all = df_all.set_index('year')
#df_all = d
```

```python
##df_men = df[df['Gender']=='M']
#df_men = df_men.drop(columns=['Gender'])
#df_men = df_men.set_index('Category')
df_men = df[['year', 'work men', 'leisure men', 'home men', 'childcare men']]
#df_men1 = df_men.drop(columns=['2020', 'Average 2017-2019'])
df_men = df_men.set_index('year')
#df_men = df_men.T
df_men
```

```python
df_women = df[['year', 'work women', 'leisure women', 'home women', 'childcare women']]
#df_men1 = df_men.drop(columns=['2020', 'Average 2017-2019'])

df_women = df_women.set_index('year')
#df_women = df_women.T
df_women
```

```python
ax = sns.lineplot(x=df_all.index, y=df_all['work paper all'], dashes=False, label='ATUS (all)')
ax = sns.lineplot(x=df_men.index, y=df_men['work men'], style=True, dashes=[(2,2)], legend=False, label='ATUS (men)')
ax = sns.lineplot(x=df_women.index, y=df_women['work women'], style=True, dashes=[(2,2)], legend=False, label='ATUS (women)')
ax.set_xticklabels(df_all.index,rotation=90)
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
ax.set(ylabel='Market Work',xlabel='Year', title='Market Work')
#ax.axvline("2008", color="black", linestyle="--")
#ax.axvline("2020", color="black", linestyle="--")
plt.savefig('fig1.pdf', bbox_inches='tight')
```

```python
df = pd.read_csv('pandemic_mkt_work.csv')
df['month'] = df['month'].astype(str)
df1 = df[df['gender']=='B']
df1 = df1.set_index('month')
df2 = df[df['gender']=='M']
df2 = df2.set_index('month')
df3 = df[df['gender']=='F']
df3 = df3.set_index('month')
```

```python
ax = sns.lineplot(x=df1.index, y=df1['coeff'], dashes=False, label='ATUS (all)')
ax = sns.lineplot(x=df2.index, y=df2['coeff'], style=True, legend=False, label='ATUS (men)')
ax = sns.lineplot(x=df3.index, y=df3['coeff'], style=True, legend=False, label='ATUS (women)')
ax = sns.lineplot(x=df1.index, y=df1['coeff_avg'], dashes=[(10,10)], label='ATUS (all) Average')
ax = sns.lineplot(x=df1.index, y=df1['lower_avg'], dashes=[(10,10)], color='lavender')
ax = sns.lineplot(x=df1.index, y=df1['upper_avg'], dashes=[(10,10)], color='lavender')
ax = sns.lineplot(x=df2.index, y=df2['coeff_avg'], dashes=[(10,10)], label='ATUS (men) Average')
ax = sns.lineplot(x=df2.index, y=df2['lower_avg'], dashes=[(10,10)], color='lavender')
ax = sns.lineplot(x=df2.index, y=df2['upper_avg'], dashes=[(10,10)], color='lavender')
ax = sns.lineplot(x=df3.index, y=df3['coeff_avg'], dashes=[(10,10)], label='ATUS (women) Average')
ax = sns.lineplot(x=df3.index, y=df3['lower_avg'], dashes=[(10,10)], color='lavender')
ax = sns.lineplot(x=df3.index, y=df3['upper_avg'], dashes=[(10,10)], color='lavender')
ax.set_xticklabels(df1.index,rotation=90)
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
plt.fill_between(df1.index, df1['lower_avg'], df1['coeff_avg'], color='lavender')
plt.fill_between(df1.index, df1['coeff_avg'], df1['upper_avg'], color='lavender')
plt.fill_between(df2.index, df2['lower_avg'], df2['coeff_avg'], color='lavender')
plt.fill_between(df2.index, df2['coeff_avg'], df2['upper_avg'], color='lavender')
plt.fill_between(df3.index, df3['lower_avg'], df3['coeff_avg'], color='lavender')
plt.fill_between(df3.index, df3['coeff_avg'], df3['upper_avg'], color='lavender')
ax.axvline("March'20", color="black", linestyle="--")
ax.set(ylabel='Market Work',xlabel='Year', title='Market Work')
#ax.axvline("2020", color="black", linestyle="--")
plt.savefig('fig5.pdf', bbox_inches='tight')
```
